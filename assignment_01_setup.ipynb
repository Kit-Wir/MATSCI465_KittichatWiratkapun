````python
import imageio.v2 as imageio
import matplotlib.pyplot as plt
import numpy as np

# Load the EM image
img = imageio.imread("example_EM_Image.tif")
img = np.array(img)

# Display the image
plt.imshow(img, cmap="gray")
plt.colorbar()
plt.title("Example EM Image")
plt.show()
````

````python
import numpy as np

mean_intensity = np.mean(img)
std_intensity = np.std(img)

print("Mean intensity:", mean_intensity)
print("Standard deviation:", std_intensity)
````

````python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(img.ravel(), bins=100)
plt.xlabel("Pixel intensity")
plt.ylabel("Count")
plt.title("Histogram of pixel intensities")
plt.show()
````

The histogram shows a broad distribution of pixel intensities, indicating significant contrast in the electron microscopy image.
Lower-intensity values correspond to darker background regions, while higher-intensity values are associated with brighter atomic-scale features.
The mean intensity represents the average image brightness, and the standard deviation reflects the overall contrast in the image.
